```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
df = pd.read_excel('us_housing_april_2010_july_2019.xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
# clean up the state
df['State'] = df['State'].apply(lambda s: s.replace('.','').strip())
df['State']
```

```python
df.head()
```

```python
df_T = df.T
df_T.columns = df['State']
df_T = df_T.drop('State')
df_T = df_T.reset_index()
# this doesn't rename the index for some reason.
df_T = df_T.rename(columns={'index':'Year'}, index={'State':''})
```

```python
#TODO: make a function out of this where input is the year, and some other parameters
# example of doing popluation per year

#2010
current_pop = df.sort_values(by=[2010],ascending=False) #df.sort_values(by=[2010],ascending=False)
plt.rcParams['figure.figsize'] = (25, 7)
ax = sns.barplot(x = current_pop['State'][:25], y = current_pop[2010][:25], palette = 'dark')
ax.set_xlabel(xlabel = 'US States', fontsize = 15)
ax.set_ylabel(ylabel = 'Housing Units in Millions', fontsize = 15)
ax.set_title(label = 'Estimate of Housing Units of US States in 2010', fontsize = 18)
ax.set_yscale('log')
plt.xticks(rotation = 90)
plt.show()

#2019
current_pop = df.sort_values(by=[2019],ascending=False) #df.sort_values(by=[2010],ascending=False)
ax = sns.barplot(x = current_pop['State'][:25], y = current_pop[2019][:25], palette = 'dark')
ax.set_xlabel(xlabel = 'US States', fontsize = 15)
ax.set_ylabel(ylabel = 'Housing Units in Millions', fontsize = 15)
ax.set_title(label = 'Estimate of Housing Units of US States in 2019', fontsize = 18)
ax.set_yscale('log')
plt.xticks(rotation = 90)
plt.show()
```
